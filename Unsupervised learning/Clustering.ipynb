#Clustering - K-Means



```python
iris_df = sns.load_dataset('iris')
iris_df
```

```python
num_cols = iris_df.select_dtypes(include=np.number).columns.tolist()
num_cols
```

```python
X = iris_df[num_cols]
X
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters =3,random_state=42)
```

```python
model.fit(X)
```

```python
model.cluster_centers_
```

```python
pred = model.predict(X)
pred
```

```python
sns.scatterplot(data=X, x='sepal_length', y='petal_length', hue=pred);
centers_x, centers_y = model.cluster_centers_[:,0], model.cluster_centers_[:,2]
plt.plot(centers_x, centers_y, 'xb')
```

```python
model =  KMeans(n_clusters=6,random_state=42)
```

```python
model.fit(X)
```

```python
pred =model.predict(X)
pred
```

```python
model.cluster_centers_
```

```python
sns.scatterplot(X,x='sepal_length',y='petal_length',hue=pred)
x_centers,y_centers= model.cluster_centers_[:,0],model.cluster_centers_[:,2]
plt.plot(x_centers,y_centers,'xb')
```

```python
model.inertia_
```

```python
inertias =[]
option = range(2,11)
for n in option:
  model = KMeans(n_clusters=n,random_state =42).fit(X)
  inertias.append(model.inertia_)
plt.plot(option,inertias,'-o')
plt.title('no.of clusters vs inertia')
plt.ylabel('inertia')
plt.ylabel('no.of clusters ')
```

# Clustering - DBSCAN

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps=0.8,min_samples=4).fit(X)
```
