```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
inputs = iris.data
targets = iris.target
```

```python
train_inputs, test_inputs, train_targets, test_targets = train_test_split(inputs, targets, random_state=42, test_size=0.25)
```

```python
len(train_inputs),len(test_inputs),len(train_targets),len(test_targets)
```

```python
model = DecisionTreeClassifier(random_state=None).fit(train_inputs,train_targets)
```

```python
from sklearn.tree import plot_tree, export_text
plt.figure(figsize=(80,20))
plot_tree(model,max_depth=2,filled=True,rounded= True,feature_names=iris.feature_names)
plt.show()
```

```python
model.max_depth
```

```python
pred = model.predict(test_inputs)
pred
```

```python
predict = model.predict(train_inputs)
predict
```

```python
train_score = accuracy_score(train_targets,predict)
```

```python
test_score = accuracy_score(test_targets,pred)
```

```python
print(f"TRAIN ACCURACY : {train_score*100}%")
print(f"TRAIN ACCURACY : {test_score*100}%")
```
