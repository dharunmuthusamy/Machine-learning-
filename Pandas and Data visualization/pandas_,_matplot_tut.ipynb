```python
gas=pd.read_csv('gas_prices.csv')
plt.figure(figsize=(10,7),dpi=200)
plt.title("Gas Price amoung various countries(inYears)")
plt.xlabel("Year")
plt.ylabel("Gas Price")
plt.xticks(gas.Year[::3])
for country in gas:
    if country!='Year':
        plt.plot(gas.Year,gas[country],'.-',label=country)
plt.legend()
plt.savefig('gas.png',dpi=200)
plt.show()
```

```python
fifa=pd.read_csv('fifa_data.csv')
fifa
```

```python
plt.style.use('ggplot')
young=fifa.loc[fifa.Age<21].count().iloc[0]
mid=fifa.loc[(fifa.Age>=21) & (fifa.Age<=24)].count().iloc[0]
exp=fifa.loc[(fifa.Age>24) & (fifa.Age<=30)].count().iloc[0]
old=fifa.loc[fifa.Age>30].count().iloc[0]
age=[young,mid,exp,old]
label=['young','mid','exp','old']
plt.pie(age,labels=label,autopct='%.2f %%')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by 'Nationality' and count occurrences
grouped_data = fifa.groupby("Nationality").size()

# Filter data for nationalities with more than 250 players
filtered_data = grouped_data[grouped_data > 250]

# Plot the data
plt.figure(figsize=(10, 6), dpi=600)
plt.plot(filtered_data, '.-', label="Count by Nationality > 250")
plt.xlabel("Nationality")
plt.ylabel("Count")
plt.title("Player Count by Nationality (More than 250 Players)")
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

```
